# Coletando os dados

## Setup

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
base_site = 'https://www.pro-football-reference.com/teams/'
team = 'nwe'
season = '2019'

url = base_site + team + '/' + season + '.htm'
url
```

```python
r = requests.get(url)
r.status_code
```

```python
html = r.content
soup = BeautifulSoup(html, 'lxml')
```

#### Exportar e salvar o arquivo HTML

```python
with open('nwe_stata2019.html', 'wb') as file:
    file.write(soup.prettify('utf-8'))
```

### Extraindo as tabelas com Pandas

```python
tables = pd.read_html(url)
```

```python
pd.set_option('display.max_columns', None)

game_schedule = tables[1]
game_schedule.head()
```

```python
# Da partida
week = game_schedule['Unnamed: 0_level_0']['Week']
day = game_schedule['Unnamed: 1_level_0']['Day']
date = game_schedule['Unnamed: 2_level_0']['Date']
hour = game_schedule['Unnamed: 3_level_0']['Unnamed: 3_level_1']
result = game_schedule['Unnamed: 5_level_0']['Unnamed: 5_level_1']
OT = game_schedule['Unnamed: 6_level_0']['OT']
wld = game_schedule['Unnamed: 7_level_0']['Rec']
where = game_schedule['Unnamed: 8_level_0']['Unnamed: 8_level_1']
opponent = game_schedule['Unnamed: 9_level_0']['Opp']
points_scored = game_schedule['Score']['Tm']
points_allowed = game_schedule['Score']['Opp']

# Estatísticas do ataque
of_first_downs = game_schedule['Offense']['1stD']
of_total_yards = game_schedule['Offense']['TotYd']
of_pass_yards = game_schedule['Offense']['PassY']
of_rush_yards = game_schedule['Offense']['RushY']
of_turnovers = game_schedule['Offense']['TO']

# Estatísticas da defesa
def_first_downs = game_schedule['Defense']['1stD']
def_total_yards = game_schedule['Defense']['TotYd']
def_pass_yards = game_schedule['Defense']['PassY']
def_rush_yards = game_schedule['Defense']['RushY']
def_turnovers = game_schedule['Defense']['TO']
```

```python
season_stats = pd.DataFrame()

season_stats['week'] = week
season_stats['day'] = day
season_stats['date'] = date
season_stats['hour'] = hour
season_stats['result'] = result
season_stats['OT'] = OT
season_stats['wld'] = wld
season_stats['where'] = where
season_stats['opponent'] = opponent
season_stats['points_scored'] = points_scored
season_stats['points_allowed'] = points_allowed

season_stats['of_first_downs'] = of_first_downs
season_stats['of_total_yards'] = of_total_yards
season_stats['of_pass_yards'] = of_pass_yards
season_stats['of_rush_yards'] = of_rush_yards
season_stats['of_turnovers'] = of_turnovers

season_stats['def_first_downs'] = def_first_downs
season_stats['def_total_yards'] = def_total_yards
season_stats['def_pass_yards'] = def_pass_yards
season_stats['def_rush_yards'] = def_rush_yards
season_stats['def_turnovers'] = def_turnovers

season_stats = season_stats.drop([17])

season_stats
```

### Deletar a Bye Week

```python
opp = season_stats['opponent']

for row in season_stats.index:
    if opp[row] == 'Bye Week':
        season_stats = season_stats.drop([row])
        break
```

### Configurando a coluna 'where' para jogos dentro e fora de casa

```python
season_stats = season_stats.replace({'where':{np.NaN : 'out', '@': 'home'}})
season_stats.head(2)
```

### Configurando a coluna 'OT', 'of_turnovers' e 'def_turnovers'

```python
# Quando o jogo não teve prorrogação ou não teve turnovers, mostrar 0 ao invés de NaN

season_stats = season_stats.replace({'OT':{np.NaN:'0'}, 'of_turnovers':{np.NaN:'0'}, 'def_turnovers':{np.NaN:'0'}})
season_stats.head(2)
```

## DataFrame final

```python
season_stats
```

### Exportando o arquivo final

```python
season_stats.to_excel('nwe_season2019_final_data.xlsx', encoding='utf-8', index=False)
season_stats.to_csv('nwe_season2019_final_data.csv', encoding='utf-8', index=False)
print('Dados exportados')
```
